```python
# how to calculate fatality rate you divide the total death by total cases


df_noworld_deathrate= data_noworld.groupby('location').total_deaths.max() / data_noworld.groupby('location').total_cases.max() * 100
```

```python
df_noworld_deathrate
```

```python
#know I want to calucate the median age 
df_age65 = data_noworld.groupby('location').aged_65_older.mean()
```

```python
df_age65
```

```python
df_age65.corr(df_noworld_deathrate, method ='pearson') ** 2
```
