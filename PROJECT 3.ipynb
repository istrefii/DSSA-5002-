```python
import pandas as pd
```

```python
import matplotlib.pylab as plt
```

```python
import numpy as np
```

```python
df = pd.read_csv('2017.csv')
```

```python
df1= pd.read_csv('2019.csv')
```

```python
df2= pd.read_csv('q1_2020.csv')
```

```python
df3= pd.read_csv('2018.csv')
```

```python
df= pd.concat([df,df1,df2,df3])
```

```python
df
```

```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time']= pd.to_datetime(df['end_time'])
```

```python
df= df.sort_values(by='start_time').reset_index(drop=True)
```

```python
df_timeindex= df.set_index('start_time')
df_timeindex.between_time('0:00', '1:00')
```

```python
fig, axes= plt.subplots(nrows=3, ncols=1,figsize=(10,20),gridspec_kw={'hspace':1}, sharey=True)

ax= axes[0]
ax.set_ylabel('January 2017')
df_timeindex[
    (df_timeindex['end_time'] >= pd.to_datetime('2017-01-01'))
    &
    (df_timeindex['end_time'] < pd.to_datetime('2017-02-01'))
].resample('1D').duration.sum().plot(kind= 'bar', ax=ax)

ax= axes[1]
ax.set_ylabel('February 2017')
df_timeindex[
    (df_timeindex['end_time'] >= pd.to_datetime('2017-02-02'))
    &
    (df_timeindex['end_time'] < pd.to_datetime('2017-03-01'))
].resample('1D').duration.sum().plot(kind= 'bar', ax=ax)

ax= axes[2]
ax.set_ylabel('March 2017')
df_timeindex[
    (df_timeindex['end_time'] >= pd.to_datetime('2017-03-02'))
    &
    (df_timeindex['end_time'] < pd.to_datetime('2017-04-01'))
].resample('1D').duration.sum().plot(kind= 'bar', ax=ax)

plt.savefig('Q1')
```
